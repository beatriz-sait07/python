```python
from matplotlib import image, pyplot
```

carrega a imagem convertida como arry numpy

```python
dt = image.imread('/content/drive/MyDrive/Colab Notebooks/projeto_ferias/Sign 0 (1).jpeg')
print(dt)
```

```python
print(dt.dtype) #visualizar o tipo de dado que foi carregado
print(dt.shape) #forma do array, tamanho dele por dimensao e qtd de tuplas
print(dt.max()) #valor max de bits
print(dt.min()) #valor min de bits
```

```python
pyplot.imshow(dt) #usando o pyplot para retornar ela de array para grafico
```

```python
img2 = Image.fromarray(dt) #convertendo de array numpy para img
type(img2)
```

voce pode usar o metodo save para salvar a nova imagem no disco, basta passar o diretorio e o formato. ex: img2.save(diretorio, format='png')
