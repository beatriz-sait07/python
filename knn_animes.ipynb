```python
def buscar_anime(nome_digitado, dataset_animes):
    resultado = []

    for anime in dataset_animes:
        if nome_digitado.lower() in anime.lower():
            resultado.append(anime)

    return resultado

# Exemplo de uso
dataset_animes = ["Naruto Next Generation: Boruto", "One Piece", "Dragon Ball Z", "Attack on Titan", "My Hero Academia"]
nome_digitado = input("Digite o nome do anime: ")

animes_encontrados = buscar_anime(nome_digitado, dataset_animes)

if animes_encontrados:
    print("Animes encontrados:")
    for anime in animes_encontrados:
        print(anime)
else:
    print("Nenhum anime encontrado com o nome digitado.")
```
